Dalam sektor perbankan, peran deposit sangat krusial. Deposit merupakan dana yang disimpan oleh nasabah di bank selama jangka waktu tertentu, dengan imbal hasil (return) karena telah meminjamkan uangnya. Bank memerlukan deposit sebagai sumber dana untuk memberikan pinjaman kepada nasabah lain dan menginvestasikan dana, yang pada akhirnya menghasilkan return yang lebih besar. Selain itu, deposit juga mendukung operasional harian bank dan menjaga likuiditas serta stabilitas keuangan. Melalui program pemasaran yang efektif, bank dapat mendorong partisipasi aktif nasabah dalam menyimpan dana melalui berbagai produk deposit yang ditawarkan. Tujuan utama program pemasaran adalah meningkatkan jumlah deposit yang masuk ke bank, yang pada gilirannya memberikan manfaat jangka panjang untuk pertumbuhan dan stabilitas keuangan bank.[sumber:financestrategist.com](https://www.financestrategists.com/banking/bank-account/bank-account-deposits/?ssp=1&darkschemeovr=1&setlang=id&cc=ID&safesearch=moderate)

## **Metric Evaluation :**

Karena fokus utama kita adalah nasabah yang berminat melakukan term deposit, maka kita menetapkan target sebagai berikut:

* 0 : Tidak melakukan term deposit
* 1 : Melakukan term deposit


Selain itu, kita perlu memahami dua jenis kesalahan yang mungkin terjadi dalam prediksi:

1. Type 1 error (False Positive): Ini terjadi ketika nasabah sebenarnya tidak akan melakukan term deposit, tetapi diprediksi akan melakukannya. Konsekuensinya adalah potensi pengeluaran biaya yang sia-sia karena Bank akan menghubungi nasabah tersebut padahal sebenarnya tidak ada minat untuk melakukan deposit.
2. Type 2 error (False Negative): Ini terjadi ketika nasabah sebenarnya akan melakukan term deposit, tetapi diprediksi tidak akan melakukannya. Konsekuensinya adalah kehilangan nasabah yang sebenarnya akan melakukan term deposit, yang dapat menambah keuntungan Bank, karena diputuskan untuk tidak menghubungi mereka berdasarkan prediksi yang salah.

Untuk memberikan gambaran secara kuantitatif, kita akan mencoba menghitung dampak biaya pada error diatas berdasarkan beberapa asumsi dibawah :

* Biaya Panggilan Telepon: Menurut [liveagent.com](https://www.liveagent.com/customer-support-glossary/cost-per-call/), biaya rata-rata untuk setiap panggilan telepon berkisar antara 2.53 EURO hingga 5.15 EURO.
* Keuntungan Term Deposit: [Investopedia.com](https://www.investopedia.com/terms/t/timedeposit.asp), menyatakan bahwa term deposit sangat menguntungkan bagi bank. Setiap deposit dari nasabah dapat digunakan oleh bank untuk meminjamkan uang ke nasabah lain atau menginvestasikannya pada sekuritas lain yang berpotensi memberikan keuntungan lebih tinggi daripada yang dibayarkan oleh nasabah.
* Frekuensi Telepon: Berdasarkan [blog.hubspot.com](https://blog.hubspot.com/sales/best-times-to-connect-with-leads-infographic), setidaknya kita perlu menelfon konsumen sebanyak 6 kali agar mereka tertarik pada produk yang kita tawarkan.
* Suku Bunga Deposit dan Lending di Portugal: Pada tahun 2014, suku bunga deposit di Portugal adalah 0.2% berdasarkan [tradingeconomics.com](https://tradingeconomics.com/portugal/interest-rate), sementara suku bunga lending berkisar sekitar 5.5% juga berdasarkan [tradingeconomics.com](https://tradingeconomics.com/portugal/bank-lending-rate).
* Deposit Berjangka: Menurut [withportugal.com](https://withportugal.com/en/immigration/depositos-portugal), deposito berjangka di Portugal diperbolehkan dengan rentang nominal antara 100 EURO hingga 250,000 EURO. Mengingat mayoritas pekerjaan dalam dataset adalah admin, dengan rata-rata gaji pertahun sekitar 18,780 EURO berdasarkan informasi dari [paylab.com](https://www.paylab.com/pt/salaryinfo/administration#:~:text=The%20gross%20salary%20range%20fordrastically%20among%20different%20job%20positions.), dan kebutuhan hidup pada tahun 2014 sekitar 12,752 EURO [numbeo.com](https://www.numbeo.com/cost-of-living/historical-data?itemId=105&city_id=6029&name_city_id=Lisbon%2C+Portugal), kita dapat mengasumsikan bahwa deposit yang dilakukan oleh satu nasabah admin adalah sekitar 60% dari sisa uangnya, yaitu 0.6 * 6028 EURO = 3617 EURO.


Berdasarkan asumsi di atas maka kita dapat coba kuantifikasi konsekuensi dari setiap type error  sebagai berikut :
- Type 1 Error (False Positive) Pengeluaran biaya --> Untuk setiap panggilan, kita asumsikan biayanya adalah 3 EURO dan kita harus menelfon sebanyak 6 x, jadi biaya yang dikeluarkan adalah 18 EURO per nasabah per tahun

- Type 2 Error (False Negative) (**Fokus untuk diturunkan untuk memperbesar nilai True Positive**) Kerugian akibat kehilangan nasabah potensial --> Untuk setiap pelanggan yang melakukan term deposit sebesar 3,617 EURO dalam jangka waktu 1 tahun, maka bank akan mendapatkan keuntungan sebesar : 
    - Keuntungan = (Nominal Deposit * (Bunga pinjaman - Bunga Deposit))
    - Keuntungan = (3617 x (5.5%-0.2%)) = 192 EURO per tahun

Berdasarkan informasi ini, kita akan membuat suatu business metric yang merupakan peningkatan keuntungan jika kita memakai machine learning yang akan kita namakan sebagai Business Metric dengan menggunakan **Recall Evaluation Metrics Method**

Menggunakan evaluation metrics recall dikarenakan recall dapat menurunkan nilai False Negative dan meningkatkan nilai True Positive. Menurut asumsi kami, konsekuensi dari False Negative lebih mengakibatkan kerugian yang lebih besar dibanding False Positive. dikarenakan False Positive merupakan sebuah Biaya Marketing yang sudah pasti dikeluarkan (Operational Cost) yang sudah dibudgetkan oleh perusahaan sehingga sudah dianggap habis.

```python
# Menambahkan pengelompokan usia untuk membantu analisis visual

def rangeumur (int) :
    if 17 <= int <= 25 :
        return '17 - 25'
    elif 25 < int <= 35 :
        return '26 - 35'
    elif 35 < int <= 45 :
        return '36 - 45'
    elif 45 < int <= 55 :
        return '46 - 55'
    elif 55 < int <= 65 :
        return '56 - 65' 
    else :
        return '> 65'
```

### **Insight**

```python
plt.figure(figsize=(20,15))
for kolom in numerical_features:

    fig, (ax1, ax2) = plt.subplots(2, sharex = True, figsize = (7.5,5), 
                                    gridspec_kw={"height_ratios": (.15, .85),
                                                "hspace": 0.05})

    stat, pvalue = normaltest(df[kolom].dropna())

    if pvalue > 0.05 :
        distribusi = f'pvalue = {pvalue:.5f}, data terdistribusi normal'
    else :
        distribusi = f'pvalue = {pvalue:.5f}, data TIDAK terdistribusi normal'

    ax1.set_title(f"Distribution of {kolom}\n{distribusi}", color = "black")
    sns.boxplot(x = df[kolom], ax = ax1)
    ax1.set(xlabel = "")

    sns.histplot(data = df[kolom], ax = ax2, stat = "count", kde=True, bins=50)

    plt.show()
```

```python
def predict_job(row):
    if row['job'] == 'housemaid':
        return (10.00 / 31.43) * 10
    elif row['job'] == 'services':
        return (8.14 / 31.43) * 10
    elif row['job'] == 'admin.':
        return (12.97 / 31.43) * 10
    elif row['job'] == 'blue-collar':
        return (6.90 / 31.43) * 10
    elif row['job'] == 'technician':
        return (10.83 / 31.43) * 10
    elif row['job'] == 'retired':
        return (25.26 / 31.43) * 10
    elif row['job'] == 'management':
        return (11.22 / 31.43) * 10
    elif row['job'] == 'self-employed':
        return (10.49 / 31.43) * 10
    elif row['job'] == 'unemployed':
        return (14.20 / 31.43) * 10
    elif row['job'] == 'entrepreneur':
        return (8.52 / 31.43) * 10
    elif row['job'] == 'student':
        return (31.43 / 31.43) * 10
    else:
        return (11.21 / 31.43) * 10
    
def predict_age(row):
    if row['age'] >= 17 and row['age'] <=25 :
        return (20.96/46.93)*10
    elif row['age'] > 25 and row['age'] <=35 :
        return (11.72/46.93)*10
    elif row['age'] > 35 and row['age'] <=45 :
        return (8.51/46.93)*10
    elif row['age'] > 45 and row['age'] <=55 :
        return (8.69/46.93)*10
    elif row['age'] > 55 and row['age'] <=65 :
        return (15.22/46.93)*10
    else:
        return (46.93/46.93)*10
    
def predict_education(row):
    if row['education'] == 'basic.4y' :
        return (10.25/22.22)*10
    elif row['education'] == 'basic.6y' :
        return (8.21/22.22)*10
    elif row['education'] == 'basic.9y' :
        return (7.82/22.22)*10
    elif row['education'] == 'high.school' :
        return (10.84/22.22)*10
    elif row['education'] == 'illiterate' :
        return (22.22/22.22)*10
    elif row['education'] == 'professional.course' :
        return (11.35/22.22)*10
    elif row['education'] == 'university.degree' :
        return (13.72/22.22)*10
    else:
        return (14.51/22.22)*10
    
def predict_marital(row):
    if row['marital'] == 'divorced' :
        return (10.32/15.00)*10
    elif row['marital'] == 'married' :
        return (10.16/15.00)*10
    elif row['marital'] == 'single' :
        return (14.01/15.00)*10
    else:
        return (15.00/15.00)*10
    
def predict_loan(row):
    if row['loan'] == 'yes' :
        return (10.93/11.34)*10
    elif row['loan'] == 'no' :
        return (11.34/11.34)*10
    else:
        return (10.81/11.34)*10
    
def predict_housing(row):
    if row['housing'] == 'yes' :
        return (11.62/11.62)*10
    elif row['housing'] == 'no' :
        return (10.88/11.62)*10
    else:
        return (10.81/11.62)*10

def predict_default(row):
    if row['housing'] == 'yes' :
        return 0
    elif row['housing'] == 'no' :
        return (12.88/12.88)*10
    else:
        return (5.15/12.88)*10
    
def predict_contact(row):
    if row['contact'] == 'celullar' :
        return (14.74/14.74)*10 
    else:
        return (5.23/14.74)*10
    
def predict_month(row):
    if row['month'] == 'mar' :
        return (50.55/50.55)*10
    elif row['month'] == 'apr' :
        return (20.49/50.55)*10
    elif row['month'] == 'may' :
        return (6.44/50.55)*10
    elif row['month'] == 'jun' :
        return (10.51/50.55)*10
    elif row['month'] == 'jul' :
        return (9.04/50.55)*10
    elif row['month'] == 'aug' :
        return (10.61/50.55)*10
    elif row['month'] == 'sep' :
        return (44.91/50.55)*10
    elif row['month'] == 'oct' :
        return (43.93/50.55)*10
    elif row['month'] == 'nov' :
        return (10.15/50.55)*10
    else:
        return (48.90/50.55)*10				

def predict_day(row):
    if row['day_of_week'] == 'mon' :
        return (9.95/12.11)*10 
    elif row['day_of_week'] == 'tue' :
        return (11.79/12.11)*10 
    elif row['day_of_week'] == 'wed' :
        return (11.67/12.11)*10 
    elif row['day_of_week'] == 'thu' :
        return (12.11/12.11)*10 
    else:
        return (10.81/12.11)*10

def predict_duration(row):
    if row['duration'] >= 1 and row['duration'] <=600 :
        return (7.84/75)*10
    elif row['duration'] > 600 and row['duration'] <=1200 :
        return (46.19/75)*10
    elif row['duration'] > 1200 and row['duration'] <=1800 :
        return (63.35/75)*10
    elif row['duration'] > 1800 and row['duration'] <=2400 :
        return (60.00/75)*10
    elif row['duration'] > 2400 and row['duration'] <=3000 :
        return (75.00/75)*10
    elif row['duration'] > 3000 and row['duration'] <=3600 :
        return (30.00/75)*10
    else:
        return (33.33/75)*10
				
def predict_pdays(row):
    if row['pdays'] == 'Never' :
        return (9.26/63.83)*10 
    else:
        return (63.83/63.83)*10

def predict_previous(row):
    if row['previous'] == 0:
        return (8.83/72.22)*10
    elif row['previous'] == 1:
        return (21.20/72.22)*10
    elif row['previous'] == 2:
        return (46.42/72.22)*10
    elif row['previous'] == 3:
        return (59.26/72.22)*10
    elif row['previous'] == 4:
        return (54.29/72.22)*10
    elif row['previous'] == 5:
        return (72.22/72.22)*10
    elif row['previous'] == 6:
        return (60/72.22)*10
    else:
        return 0
    
def predict_poutcome(row):
    if row['poutcome'] == 'success' :
        return (65.11/65.11)*10 
    elif row['poutcome'] == 'failure' :
        return (14.23/65.11)*10 
    else:
        return (8.83/65.11)*10


def predict(row):
    job_prediction = predict_job(row)
    age_prediction = predict_age(row)
    education_prediction = predict_education(row)
    marital_prediction = predict_marital(row)
    loan_prediction = predict_loan(row)
    housing_prediction = predict_housing(row)
    default_prediction = predict_default(row)
    contact_prediction = predict_contact(row)
    month_prediction = predict_month(row)
    day_prediction = predict_day(row)
    duration_prediction = predict_duration(row)
    pdays_prediction = predict_pdays(row)
    previous_prediction = predict_previous(row)
    poutcome_prediction = predict_poutcome(row)
    return (job_prediction + age_prediction + education_prediction + marital_prediction + loan_prediction + housing_prediction + default_prediction + contact_prediction + month_prediction + day_prediction + duration_prediction + pdays_prediction + previous_prediction + poutcome_prediction) / 14

# Apply the prediction function to the DataFrame
df_nomodel['predicted_y'] = df_nomodel.apply(predict, axis=1)

# Convert the continuous predictions to binary using a threshold
threshold = 5.0
df_nomodel['predicted_y_binary'] = df_nomodel['predicted_y'].apply(lambda x: 1 if x >= threshold else 0)

# Print the DataFrame with actual and predicted target values
print(df_nomodel[['Target', 'predicted_y', 'predicted_y_binary']])

# Calculate precision, recall, and accuracy
precision_nomodel = precision_score(df_nomodel['Target'], df_nomodel['predicted_y_binary'])
recall_nomodel = recall_score(df_nomodel['Target'], df_nomodel['predicted_y_binary'])
accuracy_nomodel = accuracy_score(df_nomodel['Target'], df_nomodel['predicted_y_binary'])

print(f'Precision: {precision_nomodel}')
print(f'Recall: {recall_nomodel}')
print(f'Accuracy: {accuracy_nomodel}')

# Create confusion matrix
conf_matrix = confusion_matrix(df_nomodel['Target'], df_nomodel['predicted_y_binary'])

# Plot confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Predicted No', 'Predicted Yes'], yticklabels=['Actual No', 'Actual Yes'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

* Akurasi yang Tinggi: Akurasi yang tinggi bisa menyesatkan jika dataset tidak seimbang (misalnya, lebih banyak 0 daripada 1). Pastikan untuk mengevaluasi model dengan metrik lain seperti F1-score, yang menggabungkan precision dan recall.

* Analisis Data Lebih Lanjut: Lakukan analisis lebih mendalam untuk memahami distribusi target dan prediksi. Periksa confusion matrix untuk mendapatkan gambaran yang lebih jelas tentang kinerja model.

## **Potensi Kerugian Kondisi False Positive dan False Negative**

Untuk menghitung dampak bisnis dari penggunaan kedua threshold ini, kita dapat mempertimbangkan beberapa faktor seperti biaya pemasaran dan potensi keuntungan dari setiap pelanggan yang terdaftar.

dengan asumsi seperti yang kita jelaskan sebelumnya bahwa biaya marketing diasumsikan 18 Euro dengan penjelasan 3Euro x 6 Panggilan. sedangkan untuk potensi keuntungan setiap pelanggan yang melakukan deposito diasumsikan 192 Euro.

**Threshold 0.5**
* Biaya Pemasaran (False Positives):
    Biaya pemasaran untuk 783 false positives.
* Potensi Kerugian (False Negatives):
    Kerugian dari 242 pelanggan yang gagal mendaftar. 
    
* Biaya Pemasaran : 783 x 18 Euro = 14,094 Euro
* Potensi kerugian: 192 x 242 = 46,464 Euro
* Kerugian : 46,464 + 14,094= 60,558 Euro


**Threshold 0.6**

* Biaya Pemasaran (False Positives):
    Biaya pemasaran untuk 596 false positives.
* Potensi kerugian (False Negatives):
    Kerugian dari 278 pelanggan yang benar-benar mendaftar. 

* Biaya Pemasaran : 596 x 18 Euro = 10,728 Euro
* Potensi Kerugian: 192 x 278 = 53,376 Euro
* Margin : 53,376 + 10,728 =  64,104 Euro


Dari perhitungan kerugian diatas, menunjukkan bahwa threshold 0.6 memiliki kerugian lebih besar dibandingan dengan threshold 0.5. jika merajuk pada EDA yang ada, dimana dari keseluruhan nasabah yang tidak pernah dihubungi 90.37% tidak melakukan deposito. Hal ini dianggap kerugian karena kehilangan kesempatan nasabah yang akan melakukan deposito di bank karena kurangnya informasi dari pihak marketing bank (False Negatives tidak difokuskan untuk dihubungi).
